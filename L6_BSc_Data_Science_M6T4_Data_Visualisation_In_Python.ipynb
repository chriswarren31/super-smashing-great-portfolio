### **1.1.1 Visualising the Titantic Dataset**

The Titanic dataset provides historical data on the passengers aboard the RMS Titanic, the British luxury liner that sank on its maiden voyage in 1912. The dataset includes various details such as age, fare, class, and survival status, making it a popular resource for data analysis and machine learning exercises.


image.png

````python
import pandas as pd

# Loading our titantic dataset CSV
titanic = pd.read_csv('https://raw.githubusercontent.com/BPP-Digital-Advanced-Data-Analytics/public_datasets/main/titanic.csv')

titanic.head()
````

````python
import matplotlib.pyplot as plt

# Get the number of passengers who survived and did not survive
# titanic['survived'] == 1: This part filters the dataset to include only the
# rows where the 'survived' column has a value of 1, meaning the passenger survived.
# shape[0]: This part tells us the number of rows in the filtered dataset, which gives us the total number of survivors.
n_survived = titanic[titanic['survived'] == 1].shape[0]
n_not_survived = titanic[titanic['survived'] == 0].shape[0]

# Create a bar chart
plt.bar(['Survived', 'Did not survive'], [n_survived, n_not_survived])

# Set the title and labels
plt.title('Titanic Survival Rates')
plt.xlabel('Survival')
plt.ylabel('Number of passengers')

# Show the plot
plt.show()
````

## **1.2 Anatomy of a Matplotlib plot**

A Matplotlib plot consists of the following components:

- **Figure**: The figure is the canvas on which the plot is drawn.
- **Axes**: The axes are the horizontal and vertical lines that define the coordinate system of the plot.
- **Ticks**: The ticks are the marks on the axes that indicate the values of the data.
- **Labels**: The labels are the text that describes the axes and ticks.
- **Title**: The title is the text that appears at the top of the plot.
- **Legend**: The legend is a box that identifies the different lines or points in the plot.

The `plt.bar()` function is used to create a bar chart. The first argument to the `plt.bar()` function is a list of strings that represent the categories on the x-axis. The second argument is a list of numbers that represent the values on the y-axis.

The `plt.title(), plt.xlabel()`, and `plt.ylabel(`) functions are used to set the title, x-axis label, and y-axis label of the plot, respectively.

The `plt.show() `function is used to display the plot.

## **1.3 Customisng Matplotlib Plots**

One of the great things about Matplotlib is the ability to extensively customise plots. In this section we will learn how to:

- Change colors and transparency
- Adjust line style and width
- Add custom labels
- Control axis ticks and limits
- Add annotations like arrows, text boxes, etc

````python
import matplotlib.pyplot as plt

# Get counts of survivors vs non-survivors
n_survived = titanic[titanic['survived'] == 1].shape[0]
n_not_survived = titanic[titanic['survived'] == 0].shape[0]

# Create bar chart
plt.bar(['Survived', 'Did not survive'], [n_survived, n_not_survived],
color=['g', 'r'], # Set colors
alpha=0.5, # Set transparency
linewidth=2) # Set bar linewidth

# Customize tick locations and labels
plt.xticks([0,1], ['Survived', 'Did not survive'], rotation=90)
plt.yticks(range(0,2000, 500))

# Add labels and title
plt.title('Titanic Survival Rates')
plt.xlabel('Survival')
plt.ylabel('Number of passengers')

# Show plot
plt.show()
````

## **2.1 Matplotlib Plotting Interfaces**

Matplotlib provides two main ways to create plots:
1. **Pyplot interface**
2. **Object-oriented interface**

We'll compare these approaches by creating a bar chart of Titanic survival rates by gender.

### **Pyplot Interface**

The pyplot interface provides simple plotting functions for common chart types.

We can create a bar chart with just a few lines:



````python
import matplotlib.pyplot as plt

females = titanic[titanic['sex']=='female']
males = titanic[titanic['sex']=='male']

f_survived = sum(females['survived']) / len(females)
m_survived = sum(males['survived']) / len(males)

plt.bar(['Female','Male'], [f_survived, m_survived])
plt.title('Survival Rates by Gender')
plt.show()
````

The pyplot interface is great for quick visualization and plotting. But it has limited customization options.

### **Using fig, ax (Object-oriented interface)**

The object-oriented interface allows full control by directly instantiating figure and axes objects.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Create figure and axes
fig, ax = plt.subplots()

# Get data for females and males
females = titanic[titanic['sex']=='female']
males = titanic[titanic['sex']=='male']

# Calculate survival rates
f_survived = sum(females['survived']) / len(females)
m_survived = sum(males['survived']) / len(males)

# Create bar chart
ax.bar(['Female','Male'], [f_survived, m_survived])

# Add title and axis labels
ax.set_title('Survival Rates by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Survival Rate')

# Set y-axis limits
ax.set_ylim(0,1)

# Add grid lines
ax.grid(axis='y', linestyle='--')

# Show the figure
fig.tight_layout()
fig.show()
````

In this object-oriented approach we first create the figure and axes objects. We then call methods on the axes to plot data, set titles, axis labels, etc. This gives full control over the chart creation and customisation.

The object-oriented interface requires more code but enables crafting a plot exactly how you want it. The pyplot approach is better for quick and simple visualization.

### **Example of the customaisations enable using fig, ax (Object-oriented interface)**

````python
import matplotlib.pyplot as plt

# Get counts of survivors vs non-survivors
n_survived = titanic[titanic['survived'] == 1].shape[0]
n_not_survived = titanic[titanic['survived'] == 0].shape[0]

# Create figure and axes
fig, ax = plt.subplots(figsize=(6,4))

# Create bar chart
ax.bar(['Survived', 'Did not survive'], [n_survived, n_not_survived],
       color=['g', 'r'],        # Set colors
       alpha=0.5,               # Transparency
       linewidth=2,             # Line width
       edgecolor='k')           # Black edges

# Customize grid lines
ax.grid(axis='y', linestyle='--')

# Customize tick locations and labels
ax.set_xticks([0,1])
ax.set_xticklabels(['Survived', 'Did not survive'], rotation=90)
ax.set_yticks(range(0,2000, 500))

# Add annotated arrow
ax.annotate('500 passengers', xy=(0,500), xytext=(0.5,450),
            arrowprops=dict(facecolor='black'))

# Add text box
ax.text(0, 1900, 'Around 1500 died',
        bbox=dict(facecolor='red', alpha=0.5))

# Add labels and title
ax.set_title('Titanic Survival Rates')
ax.set_xlabel('Survival')
ax.set_ylabel('Number of passengers')

# Show plot
plt.show()
````

## **2.2 Introducing Subplots and Plotting Interfaces**

Matplotlib offers two main types of plotting interfaces: the pyplot interface, which provides a simple way to add plots, text, and lines to the current axes in an easy-to-use manner; and the object-oriented interface, which gives you full control over creating and customising figures and axes. In addition to single plots, Matplotlib enables you to create subplots within a figure, allowing for complex layouts and multiple charts in one display.

````python
# Survival by gender
females = titanic[titanic['sex'] == 'female']
males = titanic[titanic['sex'] == 'male']

f_survived = sum(females['survived']) / females.shape[0]
m_survived = sum(males['survived']) / males.shape[0]

# Plot survival rates
plt.bar(['Female','Male'], [f_survived, m_survived])
plt.title('Survival Rates By Gender')
````

````python
# Survival by class
class1 = titanic[titanic['pclass'] == 1]
class2 = titanic[titanic['pclass'] == 2]
class3 = titanic[titanic['pclass'] == 3]

c1_survived = sum(class1['survived']) / class1.shape[0]
c2_survived = sum(class2['survived']) / class2.shape[0]
c3_survived = sum(class3['survived']) / class3.shape[0]

# Plot survival rates
plt.bar(['1st Class','2nd Class','3rd Class'], [c1_survived, c2_survived, c3_survived])
plt.title('Survival Rates By Class')
````

## **Now let's combine them into a single plot (side by side)**

````python
# Create figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Plot gender survival
ax1.bar(['Female','Male'], [f_survived, m_survived])
ax1.set_title('Survival Rates By Gender')

# Plot class survival
ax2.bar(['1st Class','2nd Class','3rd Class'],
       [c1_survived, c2_survived, c3_survived])
ax2.set_title('Survival Rates By Class')

# Overall title
fig.suptitle('Titanic Survival', fontsize=16)

plt.show()
````

#### **Explanation of Subplots with Matplotlib**

- `plt.subplots(1, 2, figsize=(10,5))`: Makes a figure with 1 row and 2 columns of plots. The size is set to 10x5 inches.
- `fig`: The entire canvas.
- `(ax1, ax2)`: Two individual plots.

- `ax1.bar(...)`: Makes a bar chart on the first plot (ax1).
- `ax2.bar(...)`: Makes a bar chart on the second plot (ax2).

### **Line Plot**
To display the number of passengers at different fare levels, you can use a line plot.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming titanic_df is your DataFrame
# Remove NaN values and sort DataFrame by 'Age'
sorted_df = titanic.dropna(subset=['age']).sort_values(by='age')

# Calculate cumulative survival rate
sorted_df['Cumulative_Survival_Rate'] = sorted_df['survived'].expanding().mean()

plt.plot(sorted_df['age'], sorted_df['Cumulative_Survival_Rate'])
plt.title('Cumulative Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Cumulative Survival Rate')
plt.grid(True)
plt.show()
````

#### **Bar Chart**
To show the survival rates by gender, a bar chart would be appropriate.

````python
# Calculate survival rate by gender
survival_by_gender = titanic.groupby('sex')['survived'].mean()

plt.bar(survival_by_gender.index, survival_by_gender)
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
````

#### **Histogram**
You can use a histogram to see the distribution of ages.

````python
# Remove NaN values from the 'Age' column
age_data = titanic['age'].dropna()

plt.hist(age_data, bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

#### **Scatter Plot**
A scatter plot could be used to identify if there's any correlation between age and fare.

````python
plt.scatter(titanic['age'], titanic['fare'])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
````

#### **Pie Chart**
You can represent the share of passengers in each class with a pie chart.

````python
# Count passengers by class
class_count = titanic['pclass'].value_counts()

plt.pie(class_count, labels=['3rd Class', '1st Class', '2nd Class'], autopct='%1.1f%%')
plt.title('Passenger Class Distribution')
plt.show()
````

### **3.3 Custom Colour Palattes**

You can specify colours for individual categories using the palette argument in Seaborn's violinplot function. Here's how you can change the colours for weekend and weekday dates in the violin plot.


In this code, the palette argument accepts a dictionary (`custom_palette`) that maps the days of the week to colours ('r' for red and 'g' for green). This way, weekend days ('Sat', 'Sun') will appear in red, and weekdays ('Thur', 'Fri') will appear in green.

````python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Set visual style
sns.set_style("whitegrid")

# Custom colour palette: Weekend (Sat, Sun) in red, Weekday (Thur, Fri) in green
custom_palette = {'Thur': 'g', 'Fri': 'g', 'Sat': 'r', 'Sun': 'r'}

# Create a stylish violin plot with custom colours
sns.violinplot(x='day', y='tip', data=tips, palette=custom_palette)

# Add title and labels
plt.title('Violin Plot of Tips by Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Tip Amount ($)', fontsize=14)

# Show the plot
plt.show()
````

### **3.4 Combining Multiple Plots**

Combining multiple plot types can add more depth to your data visualisation. In Seaborn, overlaying a swarm plot on top of a violin plot provides a detailed view of both the distribution and individual data points.

````python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Quick data exploration
print(tips.head())

# Set visual style
sns.set_style("whitegrid")

# Custom colour palette: Weekend (Sat, Sun) in red, Weekday (Thur, Fri) in green
custom_palette = {'Thur': 'g', 'Fri': 'g', 'Sat': 'r', 'Sun': 'r'}

# Create a violin plot with custom colours
sns.violinplot(x='day', y='tip', data=tips, palette=custom_palette, inner=None)

# Overlay a swarm plot
sns.swarmplot(x='day', y='tip', data=tips, color='k', alpha=0.6)

# Add title and labels
plt.title('Violin and Swarm Plot of Tips by Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Tip Amount ($)', fontsize=14)

# Show the plot
plt.show()
````

- **Violin Plot:** Created using sns.violinplot(). We use the inner=None parameter to remove the inner bars within the violins.

- **Swarm Plot**: Created using sns.swarmplot(). This adds individual data points on top of the violin plot. The color='k' argument sets the colour of the swarm plot to black, and alpha=0.6 makes the points slightly transparent.

- **Overlaying**: Both the violin and swarm plots share the same x and y axes, so they are automatically overlaid when plotted in sequence.

- **Palette**: We use the custom_palette dictionary to control the colours of the violin plot, making weekends red and weekdays green.

## **4.1 Let's create a Pairplot**

````python
# Import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a pair plot
sns.pairplot(tips, hue='sex')

# Display the plot
plt.show()
````

### **Explanation of Results:**

In the generated pair plot, you'll see scatter plots and histograms:

- **Scatter Plots:** Each scatter plot shows the relationship between two numerical variables in the dataset. For instance, the scatter plot for "total_bill" vs "tip" helps you visually understand how the tip amount varies with the total bill.

- **Kernel Density Estimation (KDE)**: The diagonal shows a KDE plot which gives a smooth curve that provides a sense of the "density" of the data at different values, essentially smoothing out the noise in the data for easier interpretation. For example, you can observe how "total_bill" or "tip" amounts are distributed.

- **Color Hue**: Points are coloured based on the "sex" column, so you can also get an idea of how the variables relate differently for different genders.

## **4.2 Now let's create a Heatmap of Correlation Matrix**

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Calculate the correlation matrix
correlation_matrix = tips.corr()

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Display the plot
plt.show()
````

Plotly excels at creating interactive 3D plots, geographical maps, and even dashboards. Below are some examples using Python code for each of these types of visualisations.

## **6.1 3D Scatter Plot**

In this 3D scatter plot, the x, y, and z axes represent GDP per capita, life expectancy, and population, respectively. The colour represents the continent, and hovering over each point shows the country's name.

````python
import plotly.express as px

# Using a built-in dataset for demonstration
df = px.data.gapminder().query("year==2007")

# Create a 3D scatter plot
fig_3d = px.scatter_3d(df, x='gdpPercap', y='lifeExp', z='pop', color='continent', hover_name='country')

# Show the plot
fig_3d.show()
````

## **6.2 Geographical Map**

In this map, each point represents a country, sized by its population and coloured by its continent. The projection="orthographic" option sets the projection style of the map.

````python
# Create a geographical map plot
fig_map = px.scatter_geo(df, locations="iso_alpha", color="continent",
                          hover_name="country", size="pop", projection="orthographic")

# Show the plot
fig_map.show()
````

2. **df**: The DataFrame containing your data.
 - `locations="iso_alpha"`: Column in the DataFrame indicating the ISO country codes.
 - `color="continent"`: Color-codes the markers based on the continent.
 - `hover_name="country"`: The name to display when hovering over a point.
 - `size="pop"`: Sizes the markers according to the population.
 - `projection="orthographic"`: Sets the projection type of the map.
3. **Show the Plot**: fig_map.show() is used to display the map.

````python
import dash
import dash_core_components as dcc
from dash import html
from dash.dependencies import Input, Output


# Initialise the Dash app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    dcc.Dropdown(
        id='dropdown-input',
        options=[{'label': i, 'value': i} for i in df['continent'].unique()],
        value='Asia'
    ),
    dcc.Graph(id='graph-output')
])

# Callback
@app.callback(
    Output('graph-output', 'figure'),
    [Input('dropdown-input', 'value')]
)
def update_graph(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    return px.scatter(filtered_df, x='gdpPercap', y='lifeExp')

# Run the app
if __name__ == '__main__':
    # app.run_server(debug=True)
    app.run_server(mode='inline', host='localhost', port=80)
````

### â“ðŸ¤” **Explore the Interactivity Options on our Dashboard:**

1. Using the drop down menu and see how the chart updates automatically.
2. Hover over points to see the regular plotly interactivity options.

### **Explanation of the Dash Dashboard Code**
- **Import Libraries:** dash, dash_core_components, and dash_html_components are imported to create the web application.

- **Initialise Dash**: dash.Dash(__name__) initiates a new Dash application.

- **Layout**: Defines the layout of the web application. It includes a Dropdown component (dcc.Dropdown) and a Graph component (dcc.Graph).

- **Callback**: The @app.callback decorator specifies that the function below it will be triggered whenever the value of the Dropdown (dropdown-input) changes. The function updates the scatter plot accordingly.

- **Run App**: app.run_server(debug=True) starts the web server, and you can view the dashboard by navigating to the given URL.
