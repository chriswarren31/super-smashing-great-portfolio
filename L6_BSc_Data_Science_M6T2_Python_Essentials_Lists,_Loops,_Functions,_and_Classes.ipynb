Control structures are essential in programming because they allow the program to make decisions.

**`if, else & elif` Statements**

Let's consider a weather app that suggests what to wear based on the temperature.

````python
temperature = int(input("Enter the current temperature in Celsius: "))

if temperature > 30:
    # Turn on AC
    print("Wear shorts and a t-shirt.")
else:
    # Turn on heating
    print("It's cold; wear a coat.")
````

❓🤔 Have you noticed different outputs when you input different temperature values?




In reality, we wouldn't turn on heating when temperatures drop below 30. We'd turn off the AC and also turn off heating for temperatures between 20 and 30 degrees celcius.

How would we code that?

````python
temperature = int(input("Enter the current temperature in Celsius: "))

if temperature > 30:
    print("Wear shorts and a t-shirt.")
elif 20 <= temperature <= 30:
    print("A long-sleeve shirt and trousers are suitable.")
else:
    print("It's cold; wear a coat.")
````

The **elif** statement becomes handy in these cases.

The `elif` ensures that only one block of code is executed, making the conditions mutually exclusive.

For instance, in a grading system, a student can't get both an A and a B if their grade is higher than 60.

See below for example:



````python
grade = int(input("Enter student's grade: "))

if grade >= 70:
  print("A")
elif grade >= 60:
  print("B")
elif grade >= 50:
  print("C")
elif grade >= 40:
  print("D")
else:
  print("Fail")
````

#**2. Lists**

Before diving into lists and dictionaries, it's essential to recall that Python offers different types of variables such as integers (int), floating-point numbers (float), and strings (str). These are suitable for holding a single piece of data.

Recall that unlike basic data types, **lists** allow us to hold multiple items in a single structure, making it easier to perform operations on them as a group.

**The features of lists include:**

- Ordered - The items have a specific, consistent order
- Indexing - You can access items using their index
- Duplicates allowed - Lists can have items with the same value
- Mutable - Lists can be altered even after their creation

image.png

Let's recap on some simple list operations:
- **append** - The append method adds an item to the end of the list.
- **insert** - The insert method adds an item at a specified index.
- **pop** - The pop method removes and returns the item at a given index (default is the last item).
- **del** - The del statement removes an item at a specific index.

Let's explore how we do these operations.

````python
# Initialise a list named 'fruits' containing two elements: 'apple' and 'banana'.
fruits = ['apple', 'banana']

# Use the 'append' method to add 'cherry' to the end of the 'fruits' list.
fruits.append('cherry')
# Print the updated list to verify that 'cherry' has been added.
print(fruits)

# Use the 'insert' method to add 'orange' at index 1 (second position) in the 'fruits' list.
fruits.insert(1, 'orange')
# Print the updated list to verify that 'orange' has been added at the correct position.
print(fruits)

# Note: This line is repeated from the previous example but has been kept for demonstration.
# Again, we insert 'orange' at index 1, shifting existing elements at and after this index to the right.
fruits.insert(1, 'orange')
# Print the updated list to verify that another 'orange' has been added at index 1.
print(fruits)

# Use the 'del' statement to delete the element at index 0 (first position) in the 'fruits' list.
del fruits[0]
# Print the updated list to verify that the first element ('apple') has been deleted.
print(fruits)
````

### **Indexing and Slicing Lists**

- **Indexing** - allows us to fetch a particular item from the list.
- **Slicing** - allows us to get a range of items from the list. Slicing is a versatile operation and you can perform many types of extractions such as:

````python
# Negative Indexing
# Get the last element of a list
print(my_list[-1])
````

````python
#Get elements from index -4 to -1
print(my_list[-4:-1])
````

````python
# Reverse the list using slicing
print(my_list[::-1])
````

## **3. Dictionaries**

A dictionary is a collection of **key-value pairs**, where each key must be unique.

Dictionaries can store various types of data: numbers, strings, lists, and even other dictionaries.

The characteristics of dictionaries include:

- **Ordered** - As of Python 3.7+, dictionaries are ordered
- **Indexed** - Items are accessed using keys
- **No duplicates** - Keys must be unique and be numbers or strings.
- **Mutable** - You can change values (mutable means that the object can be changed after its created)

#### **Let's create a dictionary**

````python
contact = {
    'name': 'John',
    'email': 'john.doe@email.com',
    'phone': '123-456-7890'
}
print(contact)
````

#### **Accessing Values in dictionaries**

To access a value associated with a key, you use the following syntax:

````python
# We use the key 'name' to access the value stored in that key-value pair
print(contact['name'])
````

````python
# Accessing using the get() method
# The get() method returns the value for the given key if it exists, and a default value otherwise.
email = contact.get('email', 'Not Available')
print(email)

# Our default value gets returned here as 'mobile' isn't a key
mobile = contact.get('mobile', 'Not Available')
print(mobile)
````

#### **Modifying Values**
You can change the value associated with a specific key in the dictionary.

````python
# Here we change the value of the 'email' key
contact['email'] = 'john.new@email.com'
print(contact)
````

#### **Adding New Pairs**
You can add new key-value pairs to the dictionary.

````python
contact['address'] = '123 Main St'
print(contact)
````

#### **Removing Pairs**

You can remove a key-value pair using the `del` keyword.

````python
del contact['phone']
print(contact)
````

## **`for` Loops**


`for` loops in Python are used for iterating over a sequence (e.g., a list, tuple, or string).

In this example, we declare a list called fruits. The for loop iterates over this list, with the variable fruit taking on the value of each element in the list, one at a time.

````python
fruits = ['apple', 'banana', 'cherry']  # List of strings
for fruit in fruits:  # `fruit` takes on the value of each element in `fruits`
    print(fruit)  # Printing the current value of `fruit`
````

#### **Using the range() function in for loops**

`range(stop)` - Generates a sequence from 0 to stop - 1

````python
for i in range(5):
  print(i)
````

`range(start, stop)` - Generates a sequence from start to stop - 1

````python
for i in range(2,7):
  print(i)
````

`range(start, stop, step)` - Generates a sequence from start to stop - 1, incrementing by step

````python
for i in range(2,10,2):
  print(i)
````

In Python, a loop is essentially a way to iterate over a sequence. A sequence could be a list, tuple, string, or even a range.

````python
for letter in "Python":
  print(letter)
````

## **`while` Loops**

In this code below, a variable count starts at zero. The while loop runs as long as the condition count < 3 is True. Inside the loop, "Hello!" is printed, and count is incremented by 1.

````python
count = 0  # Initialize counter
while count < 3:  # Loop condition
    print("Hello!")  # Code to execute
    count += 1  # Increment counter
````

## **`break`**

The `break` statement will exit the loop entirely, skipping any subsequent iterations and statements.

````python
for number in range(5):  # Loop from 0 to 4
    if number == 3:  # If condition is met, exit the loop
        break
    print(number)  # Otherwise, print the number
````

## **`continue`**
The continue statement, on the other hand, only skips the current iteration of the loop and continues with the next iteration.

````python
for number in range(5):  # Loop from 0 to 4
    if number == 2:  # If condition is met, skip this iteration
        continue
    print(number)  # Otherwise, print the number
````

## **`enumerate`**
The `enumerate()` function returns both the index and the value of each item in the list, useful when you need to know the position of the items.

````python
for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")
````

## **`List Comprehensions`**
List comprehensions allow for concise creation of lists. In the following example, we create a list of squares for numbers ranging from 0 to 4.

````python
squares = [x ** 2 for x in range(5)]
squares
````

## **Example: Nested for Loop to Print a Multiplication Table Grid**

Let's consider an example where we use nested for loops to print a multiplication table grid for numbers 1 to 3. The outer loop will iterate through rows, and the inner loop will iterate through columns.

````python
# Outer loop to iterate through rows
for i in range(1, 4):  # i takes on values 1, 2, 3
    # Inner loop to iterate through columns
    for j in range(1, 4):  # j takes on values 1, 2, 3
        print(f"{i} x {j} = {i * j}", end='\t')  # Compute and print product, followed by a tab
    print()  # Move to the next line after completing a row
````

# **5. Functions**

Functions in Python are self-contained blocks of code that encapsulate a specific task. Functions allow for code reuse, making your programs shorter, easier to read, and easier to maintain. But functions are more than just reusable blocks of code - they offer additional advantages as follows:

1. **Code re-usability** - No need to write the same code again and again
2. **Modularity** - Breaks down complex tasks into smaller, manageable pieces
3. **Easier Debugging** - Easier to spot and correct errors in a function than in a lengthy code block



**Defining a Function**

The syntax for declaring a function is:



```
def function_name(parameters):
    # Body of the function
```

- `def` is a keyword that informs Python that you are about to define a function.
- `function_name` is the name you want to give to the function.
- `parameters` are the variables that a function receives, allowing you to pass information into the function.


**A Simple Function**

Here, we declare a function named say_hello that prints "Hello, World!" when called.

````python
def say_hello():
    print("Hello, World!")

# Calling the function
say_hello()
````

**Function with a Parameter**

Here's a function that takes a name parameter and prints a greeting.

````python
def greet(name):
    print(f"Hello, {name}!")

# Calling the function with different arguments
greet("Alice")
greet("Bob")
````

**Function that Returns a Value**

Functions can also return values back to the code that called them using the return statement.

Example: Function that Returns the Square of a Number

````python
def square(num):
    return num * num

# Storing the returned value in a variable
result = square(4)
print(result)  # Output: 16
````

# **6. Classes**
A class serves as a blueprint for creating objects, which are essentially instances of the class. A class can contain data members (attributes) and methods (functions) that operate on the attributes.

**Syntax for Defining a Class**
The basic syntax for defining a class is as follows:



```
class ClassName:
    # Class attributes and methods
```

- `class` is a keyword that signals Python you're defining a class.
- `ClassName` is the name you want to give to the class.



**A Simple Class**

Let's define a simple class called Dog with one attribute name and one method bark.

````python
class Dog:
    def __init__(self, name):
        self.name = name  # Attribute

    def bark(self):  # Method
        print(f"{self.name} says Woof!")
````

- `__init__` is a special method called a constructor. It's called when you create a new object from the class.
- `self` refers to the instance of the class. It must be the first parameter of any method in the class.

**Creating an Object**

You can create an object (an instance of a class) as follows:

````python
my_dog = Dog("Harrison")
````

Here, my_dog is an object of the class Dog, and "Fido" is passed as an argument to the __init__ method.

````python
my_dog.bark()  # Output: Fido says Woof!
````

#### **Class with Multiple Methods**
Let's expand our Dog class to include a sit method.

````python
class Dog:
    def __init__(self, name):
        self.name = name  # Attribute

    def bark(self):  # Method
        print(f"{self.name} says Woof!")

    def sit(self):  # Another method
        print(f"{self.name} is sitting.")
````

````python
# To call the new method
my_dog = Dog("Harley")
my_dog.sit()  # Output: Fido is sitting.
````
